## Lab 04 - Simple Linear Regression Scratch

### Import The Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing The Dataset

```python
# Load dataset
data = pd.read_csv("datasets/advertising.csv")
```

### Importing The Data

```python
X = data["TV"].values
y = data["Sales"].values
```

### Mean

```python
# Mean
mean_x = np.mean(X)
mean_y = np.mean(y)
```

### Calculate Slope and Intercept

```python
# Calculate slope (m) and intercept (c)
num = np.sum((X - mean_x) * (y - mean_y))
den = np.sum((X - mean_x)**2)
m = num / den
c = mean_y - m * mean_x
print("Slope (m):", m)
print("Intercept (c):", c)
```

### Prediction

```python
# Predict
y_pred = m * X + c
```

### Mean Square Error

```python
#mse
mse = np.mean((y-y_pred)**2)
print("MSE:", mse)
```

### Accuracy

```python
#accuracy
ss_total = np.sum((y-np.mean(y))**2)
ss_res = np.sum((y-y_pred)**2)
r2 = 1 - (ss_res / ss_total)
print("r2:", r2)
```

### Visaulization

```python
# Visualization
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()
```
