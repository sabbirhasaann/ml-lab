```python
import numpy as np
from sklearn.linear_model import Perceptron 
X = np.array([[0,0],[0,1],[1,0],[1,1]]) 
Y = np.array([0,1,1,1]) 
clf = Perceptron(max_iter=1000, eta0=0.1) 
clf.fit(X, Y) 
print("Weights:", clf.coef_) 
print("Bias:", clf.intercept_) 
print("Predictions:", clf.predict(X)) 
```
