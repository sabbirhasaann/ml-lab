## Import The Libraries

```python
from ydata_profiling import ProfileReport
import pandas as pd
```

## Report Function

```python
def generate_eda_report(df, output_path="EDA_Report.html", title="Automated EDA Report"):
    """
    Generates an auto EDA HTML report for any dataset using ydata-profiling.

    Parameters:
    df (DataFrame): Input DataFrame.
    output_path (str): File name/path for saving the report.
    title (str): Title of the report.

    Returns:
    None
    """

    profile = ProfileReport(
        df,
        title=title,
        explorative=True,   # richer analysis
        minimal=False,
        # correlations={
        #     "pearson": True,
        #     "spearman": True,
        #     "kendall": True,
        #     "phi_k": True,
        #     "cramers": True,
        # },
    )

    profile.to_file(output_path)
    print(f" EDA Report Generated Successfully: {output_path}")
```

## Usage Example

```python
# df = pd.read_csv("your_dataset.csv")
df = pd.read_excel("Bangladeshi_Student_Database.xlsx")

# Generate report
generate_eda_report(df, "My_EDA_Report.html")
```
